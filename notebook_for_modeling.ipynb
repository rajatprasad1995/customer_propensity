```python
from sklearn.linear_model import LogisticRegression
```

```python
def auc(variables, target, df):
    X = df[variables]
    y = df[target]
    logreg = LogisticRegression()
    logreg.fit(X, y)
    predictions = logreg.predict_proba(X)[:,1]
    auc = roc_auc_score(y, predictions)
    return(auc)
```

```python
def next_best(current_variables,candidate_variables, target, df):
    best_auc = -1
    best_variable = None
    for v in candidate_variables:
        auc_v = auc(current_variables + [v], target, df)
        if auc_v >= best_auc:
            best_auc = auc_v
            best_variable = v
    return best_variable
```

```python
def profit(perc_targets, perc_selected, population_size,reward_target,cost_campaign):
    cost = cost_campaign * perc_selected * population_size
    reward = reward_target * perc_targets * perc_selected * population_size
    return(reward - cost)
```
